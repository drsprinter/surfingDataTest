```python
import tensorflow as tf
import pandas as pd
import numpy as np
import tflearn
```

```python
df=pd.read_csv('surfData/surfData.csv',index_col=0)
#print(df)
```

```python
a=[]
for i in range(630):
    a.append([df.iat[i,18]])
print(a)
```

```python
arr=np.array(a)
aTest=[]
for i in range(630):
    if arr[i] >= 1000:
        aTest.append([1,0])
    else:
        aTest.append([0,1])
trainY=np.array(aTest)
print(trainY)
```

```python
b=[]
for i in range(630):
    b.append([df.iat[i,9],df.iat[i,11],df.iat[i,13],df.iat[i,15],df.iat[i,17]])
trainX=np.array(b)
print(trainX)
#9,weather id
#11,wind direction id
#13,wind strength id
#15,wind type id
#17,wave size id
#31,surface temperature
```

```python
c=[]
for i in range(630,640):
    c.append([df.iat[i,18]])
cArr=np.array(c)
cTest=[]
for i in range(10):
    if cArr[i]>=1000:
        cTest.append([1,0])
    else:
        cTest.append([0,1])
testY=np.array(cTest);
print(testY);
```

```python
d=[]
for i in range(630,640):
    d.append([df.iat[i,9],df.iat[i,11],df.iat[i,13],df.iat[i,15],df.iat[i,17]])
testX=np.array(d)
print(testX)
```

```python
tf.reset_default_graph()
tnorm = tflearn.initializations.uniform(minval=-1.0, maxval=1.0)
net=tflearn.input_data(shape=[None,5])
net=tflearn.fully_connected(net,4,activation='sigmoid')
net=tflearn.fully_connected(net,4,activation='sigmoid')
#net=tflearn.fully_connected(net,2,activation='sigmoid')
net = tflearn.fully_connected(net,2,activation='sigmoid')
net = tflearn.regression(net,optimizer='sgd',learning_rate=0.5,loss='mean_square')
```

```python
model = tflearn.DNN(net) 
model.fit(trainX,trainY,n_epoch=350,batch_size=256,validation_set=0.1,show_metric=True)
```

```python
pred = np.array(model.predict(testX))
print(pred)

label = testY
print(label)
```
